```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(img):
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
cars_dir = 'examples/cars/'

img = cv2.imread(cars_dir + '1.jpg')

# convert to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


show_image(img)
```
